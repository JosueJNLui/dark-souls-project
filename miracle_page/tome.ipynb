```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tabulate import tabulate
import re  # Importa o módulo de expressões regulares

def extrair_dados_tome(soup_categoria):
    try:
        data = []
        
        # Verifica se a tabela existe na página
        table = soup_categoria.find('div', class_='table-responsive')
        if not table:
            raise Exception("Tabela não encontrada na página.")

        for row in table.find('table').find_all('tr')[1:]:
            columns = row.find_all('td')
            # Adiciona verificação para garantir que há colunas suficientes
            if len(columns) >= 4:
                icon = columns[0].find('img')['src']
                tome = columns[1].get_text(strip=True)
                usage = columns[2].get_text(strip=True)
                location = columns[3].get_text(strip=True)

                data.append({
                    'Icon': icon,
                    'Tome': tome,
                    'Usage': usage,
                    'Location': location,
                })

        tome_df = pd.DataFrame(data)

        return tome_df

    except Exception as e:
        print(f"Erro ao extrair dados da categoria Tome: {e}")
        return None

# URL da categoria "tome"
url_tome = "https://darksouls3.wiki.fextralife.com/Miracles"

# Faz a requisição HTTP para a página da categoria
response_tome = requests.get(url_tome)

if response_tome.status_code == 200:
    html_tome = response_tome.text
    soup_tome = BeautifulSoup(html_tome, 'html.parser')  # Use o parser padrão aqui

    # Extrai os dados da categoria "tome"
    tome_df = extrair_dados_tome(soup_tome)

    if tome_df is not None and not tome_df.empty:
        print("Dados da categoria Tome:")
        # Exibe os dados como tabela
        print(tabulate(tome_df, headers='keys', tablefmt='pretty', showindex=False))
    else:
        print("Falha ao extrair dados específicos da categoria tome ou o DataFrame está vazio.")
else:
    print(f"Falha ao obter a página da categoria. Código de status: {response_tome.status_code}")
```
